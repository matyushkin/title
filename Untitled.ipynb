```python
import pandas as pd
from tqdm import tqdm, tqdm_notebook
tqdm_notebook().pandas()
df_meta = pd.read_feather(f'/home/leo/Yandex.Disk/GitHub/DATASETS/META/it_articles_ru.feather')
```

```python
def corr(x):
    L = df_meta[df_meta.filename == x]['index'].to_list()
    if L:
        return L[0]
    else:
        return pd.NA

def clean_fulltexts(name):
    df_fulltexts = pd.read_feather(f'/home/leo/Yandex.Disk/GitHub/DATASETS/FULLTEXTS/{name}.feather')
    print(df_fulltexts.shape)
    _ = df_fulltexts.filename.progress_apply(corr)
    df_fulltexts['filename'] = _
    df_fulltexts = df_fulltexts.dropna()
    print(df_fulltexts.shape)
    return df_fulltexts
```
